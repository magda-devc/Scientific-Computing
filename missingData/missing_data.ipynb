```python
import pandas as pd
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt

#Loading data
dat = pd.read_csv("Age-standardizedSuicideRates.csv")
print(dat)
```

```python
#Checking the information
dat.info()
```

```python
#The first 5 elemnets
dat.head()
```

```python
#The last 5 elements
dat.tail()
```

```python
#Size of data
dat.shape
# 69 rows, 6 columns
```

```python
#Summary
dat.describe(include='all')
```

```python
#total number of missing values in each column
dat.isnull().sum()
#gives the total numeber of missing values
#dat.isnull().sum().sum()
```

```python
#Handling missing values
dat.fillna(dat.mean(numeric_only=True), inplace=True)
print(dat)
```

```python
#Key statistics
dat.mode()
```

```python
#Histogram
dat.hist()
```

```python
#Specifying for one histogram
dat[[ "2000",]].hist()
```

```python
# Scatter plot
sns.scatterplot(data=dat, x="2016", y="2010", color="black")

plt.title("2010 vs 2016  scatter plot")  
plt.show()
```

```python
sns.scatterplot(data=dat, x="2015", y="Country", color="green")

plt.title("Country vs 2015  scatter plot")  
plt.show()
```

```python
#Correlation matrix

#Removing values that are not numbers
correlation_matrix = dat.select_dtypes(include=["number"]).corr()



plt.figure(figsize=(10, 6))  
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f", linewidths=0.5)

plt.title("Correlation Heatmap")
plt.show()
```
